```python
# Daily temperatures recorded in a city for a month
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
data=pd.read_excel("/content/Book1.xlsx",sheet_name=0)
days=data['DAY']
temp=data['TEMPERATURE']
temp1=list(temp)
a=np.array(temp)
mean=round(a.mean())
t_temp,t_days,n_days,n_temp,mean_days,mean_temp=[],[],[],[],[],[]
for i in range(0,len(days)):
  if(temp[i]>mean):
    t_temp.append(temp[i])
    t_days.append(days[i])
  elif(temp[i]<mean):
    n_days.append(days[i])
    n_temp.append(temp[i])
  else:
    mean_days.append(days[i])
    mean_temp.append(temp[i])
plt.scatter(t_days,t_temp,color='red',label='>threshold')
plt.scatter(n_days,n_temp,color='orange',label='<threshold')
plt.plot(mean_days,mean_temp,color='green',label='mean')
plt.scatter(mean_days,mean_temp,color='green')
plt.plot(days,temp,color='black')
plt.legend(loc='best')
plt.show()
days_counts=[len(t_days),len(n_days),len(mean_days)]
color=['red','orange','green']
label=[f'No.of days greater than threshold ({days_counts[0]})',f'No.of days less than threshold ({days_counts[1]})',f'No.of days with avg temperature ({days_counts[2]})']
plt.pie(days_counts,labels=label,colors=color)
plt.show()
x=[a.min(),a.max()]
y=[temp1.index(x[0]),temp1.index(x[1])]
plt.bar(y,x)
plt.xlabel('DAY')
plt.ylabel('TEMPERATURE')
plt.title("Minimum and Maximum temperature")
plt.show()
```

```python
from sklearn.model_selection import train_test_split
x=data_merge["class"]
y=data_merge["text"]
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.25,random_state=0)
```

```python
from sklearn.linear_model import LogisticRegression
import numpy as np
from sklearn.metrics import accuracy_score, classification_report

x_train = x_train.values.reshape(-1, 1)
x_test = x_test.values.reshape(-1, 1)

LR=LogisticRegression()
LR.fit(x_train,y_train)
accuracy = LR.score(x_test,y_test).round(2)
print("Accuracy is" , accuracy)
pred_LR= LR.predict(x_test)
print(classification_report(y_test,pred_LR))

```
