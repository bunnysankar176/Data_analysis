Extracting HTML code of any website using BeautifulSoup

```python
from bs4 import BeautifulSoup
import requests

url='http://www.w3.org/1999/xhtml'
page=requests.get(url)
htmlcode=BeautifulSoup(page.text,'html')
print(htmlcode)
```

```python
t=htmlcode.find('documentation')
print(t)
```

```python
print(t.text)
```


Extracting HTML code of any website using BeautifulSoup and saving it as.csv file

```python
from bs4 import BeautifulSoup
import requests

url='https://www.forbesindia.com/article/explainers/top-10-richest-people-india/85909/1'
page=requests.get(url)
htmlcode=BeautifulSoup(page.text,'html')
print(htmlcode)
```

```python
t=htmlcode.find('table')
print(t)
```

```python
from bs4 import BeautifulSoup
headings=t.find_all('th')
print(headings)
```

```python
head=[]
for i in headings:
  head.append(i.text)
print(head)
```

```python
#List comprehension
head =[i.text for i in headings]
print(head)
```

```python
#Creating dataframe
import pandas as pd
df = pd.DataFrame(columns=head)
print(df)
```

```python
rows=t.find_all('tr')
print(rows)
```

```python
for i in rows[1:6]:
  #print(i)
  erd=i.find_all('td')
  print(erd)
```

```python
for i in rows[1:6]:
  erd=i.find_all('td')
  rd=[j.text for j in erd]
  print(rd)
```

```python
ind=0
for i in rows[1:6]:
  erd=i.find_all('td')
  rd=[i.text for i in erd]
  print(rd)
  df.loc[ind]=rd
  ind=ind+1
df
```

```python
df.to_csv("webscrapping.csv")
```
